```python
#Loaidng required packages
import wbgapi as wb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Pandas function to perform the dataset loading
dtaframe=pd.read_csv(r"D:\\Imran\Climate_Indicators.csv", low_memory=False)
```

```python
#Function for setting the Index
dtaframe1=dtaframe.set_index('economy')
```

```python
#Showing the transpose form of the indicators
dtaframe1.T.head(12)
```

```python
#Showing new indicators related to climate and economy
indctrs_econ = ['NY.GDP.MKTP.PP.CD','SL.UEM.1524.FE.ZS','NE.DAB.TOTL.ZS','SL.UEM.1524.NE.ZS']
indctrs_clm=['EN.ATM.GHGT.KT.CE','EN.ATM.CO2E.GF.KT','EG.ELC.NGAS.ZS','EG.ELC.ACCS.RU.ZS']
codes = ['ESP','BGD','AUS','GBR','BMU','JAM','CHN','JPN']
ECN  = wb.data.DataFrame(indctrs_econ, codes, mrv=7)
CLM = wb.data.DataFrame(indctrs_clm, codes, mrv=7)

#NY.GDP.MKTP.PP.CD: Current GDP, PPP of a country
#SL.UEM.1524.FE.ZS: Female unemployment for youths in a country
#NE.DAB.TOTL.ZS: Expenditure of a country
#SL.UEM.1524.NE.ZS: Unemployment for youths in a country
#EN.ATM.GHGT.KT.CE: Greenhouse gas emissions in a country
#EN.ATM.CO2E.GF.Kt: Gaseous fuel consumption in terms of CO2 emissions 
#EG.ELC.NGAS.ZS: Total % of electricity production from natural gas sources in a country
#EG.ELC.ACCS.RU.ZS: Electricity access for a country in rural places
```

```python
# Economic indicators
ECN.columns = [x.replace('YR','') for x in ECN.columns]      
ECN=ECN.stack().unstack(level=1)                             
ECN.index.names = ['Country', 'Year']                           
ECN.columns                                                     
ECN.fillna(0)
ECN.head(5)
```

```python
# Climatic indicators
CLM.columns = [y.replace('YR','') for y in CLM.columns]      
CLM=CLM.stack().unstack(level=1)                             
CLM.index.names = ['Country', 'Year']                           
CLM.columns                                                     
CLM.fillna(0)
CLM.head(7)
```

```python
#Dataframe index reset and elimating null values
ecn1=ECN.reset_index()
clm1=CLM.reset_index()
ecn2=ecn1.fillna(0)
clm2=clm1.fillna(0)
```

```python
#Final dataframe creation
fnal = pd.merge(ecn2, clm2)
fnal.head(7)
```

```python
# Statistical analysis for Spain
cn1=fnal[(fnal['Country']=='ESP')]
cn1.describe()
```

**The average total expenditure in Spain is 97.04. The average total electricity access for a country in rural places is 100%**
